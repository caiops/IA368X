# Analisando as *features* selecionadas

```python
import pandas as pd
import numpy as np

scenario01 = pd.read_csv('https://github.com/caiops/IA368X/raw/main/p2/data/interim/features/features_observations_01.csv')
scenario02 = pd.read_csv('https://github.com/caiops/IA368X/raw/main/p2/data/interim/features/features_observations_02.csv')

display(scenario01)
display(scenario02)
```

## Morte dos pacientes

```python
print('Porcentagem de pacientes que morreram (cenário 01):')
print(scenario01['DEATH_FN'].value_counts()[1]/len(scenario01['DEATH_FN']))
print('Porcentagem de pacientes que morreram (cenário 02):')
print(scenario02['DEATH_FN'].value_counts()[1]/len(scenario02['DEATH_FN']))
```

## Sexo dos pacientes

```python
print('Sexo dos pacientes (cenário 01):')
print(scenario01['GENDER'].value_counts())
print(scenario01['GENDER'].value_counts(normalize = True))

print('\nSexo dos pacientes (cenário 02):')
print(scenario02['GENDER'].value_counts())
print(scenario02['GENDER'].value_counts(normalize = True))
```

```python
print('Sexo dos pacientes que morreram (cenário 01):')
print(scenario01.loc[scenario01['DEATH_FN'] == 1,'GENDER'].value_counts())
print(scenario01.loc[scenario01['DEATH_FN'] == 1,'GENDER'].value_counts(normalize = True))

print('\nSexo dos pacientes que não morreram (cenário 01):')
print(scenario01.loc[scenario01['DEATH_FN'] == 0,'GENDER'].value_counts())
print(scenario01.loc[scenario01['DEATH_FN'] == 0,'GENDER'].value_counts(normalize = True))

print('\nSexo dos pacientes que morreram (cenário 02):')
print(scenario02.loc[scenario02['DEATH_FN'] == 1,'GENDER'].value_counts())
print(scenario02.loc[scenario02['DEATH_FN'] == 1,'GENDER'].value_counts(normalize = True))

print('\nSexo dos pacientes que não morreram (cenário 02):')
print(scenario02.loc[scenario02['DEATH_FN'] == 0,'GENDER'].value_counts())
print(scenario02.loc[scenario02['DEATH_FN'] == 0,'GENDER'].value_counts(normalize = True))
```

```python
print('Porcentagem de mulheres que morreram (cenário 01):')
print(scenario01.loc[scenario01['DEATH_FN'] == 1,'GENDER'].value_counts()[1]/scenario01['GENDER'].value_counts()[1])
print('\nPorcentagem de homens que morreram (cenário 01):')
print(scenario01.loc[scenario01['DEATH_FN'] == 1,'GENDER'].value_counts()[0]/scenario01['GENDER'].value_counts()[0])
print('\nPorcentagem de mulheres que morreram (cenário 02):')
print(scenario02.loc[scenario02['DEATH_FN'] == 1,'GENDER'].value_counts()[0]/scenario02['GENDER'].value_counts()[0])
print('\nPorcentagem de homens que morreram (cenário 02):')
print(scenario02.loc[scenario02['DEATH_FN'] == 1,'GENDER'].value_counts()[1]/scenario02['GENDER'].value_counts()[1])
```

# Gerando as tabelas finais de *features*

```python
#Descartando a coluna com a data de nascimento
scenario01 = scenario01.drop('BIRTHDATE', axis=1)
scenario02 = scenario02.drop('BIRTHDATE', axis=1)

display(scenario01)
display(scenario02)
```

```python
#Juntando as duas tabelas para gerar uma única tabela com os dois cenários
all = pd.concat([scenario01, scenario02], ignore_index=True)
display(all)
```

```python
#Salvando as tabelas como .csv
scenario01.to_csv('scenario01.csv', index = False)
scenario02.to_csv('scenario02.csv', index = False)
all.to_csv('all.csv', index = False)
```
