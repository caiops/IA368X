# Transformação e processamento dos dados para que sejam utilizados pelo Cytoscape

```python
import pandas as pd
import numpy as np

#Leitura dos dados
raw_odds = pd.read_csv('https://github.com/aaronab/comorbidity_networks/raw/master/comorbidity_odds_matrix.csv')
raw_pmat = pd.read_csv('https://github.com/aaronab/comorbidity_networks/raw/master/comorbidity_pmat_matrix.csv')
raw_sex_odds = pd.read_csv('https://github.com/aaronab/comorbidity_networks/raw/master/comorbidity_sexeffect_odds_matrix.csv')
raw_sex_pmat = pd.read_csv('https://github.com/aaronab/comorbidity_networks/raw/master/comorbidity_sexeffect_pval_matrix.csv')
```

```python
#Visualização dos dados
display(raw_odds)
display(raw_pmat)
display(raw_sex_odds)
display(raw_sex_pmat)
```

Achatando as tabelas para o formato utilizado no Cytoscape.

```python
#Transformando a primeira coluna no index
odds = raw_odds.copy()
odds.set_index('Unnamed: 0', inplace = True)

pmat = raw_pmat.copy()
pmat.set_index('Unnamed: 0', inplace = True)

sex_odds = raw_sex_odds.copy()
sex_odds.set_index('Unnamed: 0', inplace = True)

sex_pmat = raw_sex_pmat.copy()
sex_pmat.set_index('Unnamed: 0', inplace = True)
```

```python
#Achatando as tabelas
odds_flatten = odds.stack().reset_index()
odds_flatten.columns = ['i', 'j', 'log_odds']
#Excluíndo os valores da diagonal principal
odds_flatten = odds_flatten[odds_flatten['i'] != odds_flatten['j']]

pmat_flatten = pmat.stack().reset_index()
pmat_flatten.columns = ['i', 'j', 'p_values']
#Excluíndo os valores da diagonal principal
pmat_flatten = pmat_flatten[pmat_flatten['i'] != pmat_flatten['j']]

sex_odds_flatten = sex_odds.stack().reset_index()
sex_odds_flatten.columns = ['i', 'j', 'log_odds']
#Excluíndo os valores iguais a zero
sex_odds_flatten = sex_odds_flatten[sex_odds_flatten['log_odds'] != 0]

sex_pmat_flatten = sex_pmat.stack().reset_index()
sex_pmat_flatten.columns = ['i', 'j', 'p_values']
#Excluíndo os valores iguais a zero
sex_pmat_flatten = sex_pmat_flatten[sex_pmat_flatten['p_values'] != 0]
```

```python
#Visualizando as tabelas resultantes
display(odds_flatten)
display(pmat_flatten)
display(sex_odds_flatten)
display(sex_pmat_flatten)
```

Podemos ver que, infelizmente, as tabelas referentes ao efeito do sexo continham apenas valores iguais a zero e essa parte da análise não pôde ser reproduzida.

Então, juntamos as tabelas de log odds e p-values para realizar a filtragem das comorbidades de interesse.

```python
#Juntando as tabelas de odds e p-values
odds_p = odds_flatten.merge(pmat_flatten, on=['i', 'j'], how='left')
odds_p
```

```python
#Pegando apenas as linhas com efeito positivo (odds > 1 ou log odds > 0) estatisticamente significativo (p < 0.05)
odds_p_i = odds_p.query("p_values < 0.05 & log_odds > 0").copy()
odds_p_i
```

Pegamos as listas de abreviações e de index das condições psiquiátricas (do arquivo "sample_code.R" no GitHub do artigo) para gerar uma tabela de referência.

```python
#Lista de abreviações dada no arquivo "sample_code.R" no GitHub do artigo
abbr = ["Her","Def","Adj","Ane","Ast","ADHD","Bac","Bil","Bur","NIn","CVD","COPD","Ulc","Coag","Coma","CIj","Dem","Dev","Art","Mou","Hea","Uri","Vei","WBC","Lip","Jaw","Ch","Ear","Ep","Eye","Ele",
                  "Fr","GIH","Gou","HA","NSC","HTN","Imm","Imp","ArI","IIn","BIj","TJ","Liv","LGI","Myc","NTJ","Gas","Nut","Wou","Ost","Bone","Con","End","GIO","HeO","InO","SInf","IjO","LRO","NSO",
                  "NuO","SkO","URO","Pan","Par","PFr","Per","Ple","Poi","RF","RIn","SIn","Spi","Bac","Spr","Sui","SuIj","Lup","Thy","UGI","Vir","Alc","Drg","Sch","MDD","BPD","PTSD","DepO","AnxO","GAD","DM","Neo","Con","Gen"]
#Lista de index das condições consideradas como psiquiátricas (do mesmo arquivo anterior) e subtraindo 1 (já que em R as linhas começam em 1 e em python começam em 0)
mod = np.array([3,6,17, 18, 27, 39, 68 ,77 ,83 ,84 ,85, 86, 87, 88, 89, 90, 91])
mod = mod-1
```

```python
#Criando um novo dataframe de referência que relaciona o nome da morbidade no GitHub com a abreviação e o subgrupo (psiquiátrica ou médica)
ref = pd.DataFrame(columns=['name', 'abbreviation', 'subgroup'])
ref['name'] = raw_odds['Unnamed: 0']
ref['abbreviation'] = abbr
ref['subgroup'] = ref.apply(lambda x: 'PSYCH' if x.name in mod else 'MEDIC', axis=1)
ref
```

```python
#Imprimindo apenas as morbidades psiquiátricas para comparação com a Tabela 2 do artigo
ref[ref['subgroup'] == 'PSYCH'].sort_values(by=['abbreviation'])
```

Então, substituímos os nomes das condições na tabela de comorbidades de interesse pelas suas respectivas abreviações.

```python
#Transformando as duas primeiras colunas da tabela de referência em um dicionário do nome do github para as abreviações
d = dict([(i,a) for i,a in zip(ref['name'], ref['abbreviation'])])
d
```

```python
#Substituindo os nomes do github pelas abreviações na tabela da rede
odds_p_i.replace({'i': d}, inplace = True)
odds_p_i.replace({'j': d}, inplace = True)
odds_p_i
```

Atrelamos cada condição (i e j) com seu respectivo subgrupo.

```python
#Pegando apenas as abreviações e seus respectivos subgrupos
ref_sub = ref[['abbreviation', 'subgroup']]
ref_sub.columns = ['i', 'subgroup']
ref_sub
```

```python
#Atrelando as morbidades (i) com seu subgrupo (psiquiátrico ou médico)
odds_p_i_sub = odds_p_i.merge(ref_sub, on=['i'], how='left')
odds_p_i_sub
```

```python
#Atrelando as morbidades (j) com seu subgrupo (psiquiátrico ou médico)
ref_sub_j = ref_sub.rename(columns={'i': 'j'})
odds_p_i_sub = odds_p_i_sub.merge(ref_sub_j, on=['j'], how='left', suffixes=('_i', '_j'))
```

```python
#Visualizando a matriz final
odds_p_i_sub
```

## Salvando as tabelas

```python
#Tabela relativa à rede completa
odds_p_i_sub[['i', 'j', 'log_odds']].to_csv('network_all.csv', index = False)
```

```python
#Tabela relativa à rede de condições psiquiátricas
odds_i_psych = odds_p_i_sub.query("subgroup_i == 'PSYCH' & subgroup_j == 'PSYCH'")
odds_i_psych[['i', 'j', 'log_odds']].to_csv('network_psych.csv', index = False)
```

```python
#Tabela relativa à rede de condições médicas
odds_i_medic = odds_p_i_sub.query("subgroup_i == 'MEDIC' & subgroup_j == 'MEDIC'")
odds_i_medic[['i', 'j', 'log_odds']].to_csv('network_medic.csv', index = False)
```

```python
#Tabela de referência, atrelando cada condição ao seu subgrupo
ref[['abbreviation', 'subgroup']].to_csv('ref_sub.csv', index = False)
```

## Análises básicas da rede de condições psiquiátricas

Verificando os valores máximo e mínimo de *log odds* dessa rede.

```python
print(odds_i_psych['log_odds'].max())
print(odds_i_psych['log_odds'].min())
```

Verificando a quantidade de arestas com _odds ratio_ maior que 1.5 (equivalente a um valor de _log odds_ maior que 0.405).

```python
odds_i_psych_1_5 = odds_i_psych[odds_i_psych['log_odds'] > 0.405]
print('Quantidade de arestas com odds ratio maior que 1.5:')
print(len(odds_i_psych_1_5))
```

Verificando quantas dessas arestas não são recíprocas, indicando que uma relação preditiva de uma condição A para uma B não necessariamente implica em uma relação recíproca da condição B para a A.

```python
#Criando uma lista de listas com base nas colunas i e j
relation_list = odds_i_psych_1_5[['i', 'j']].values.tolist()
#Visualizando
relation_list
```

```python
cont = 0
#Verificando se para a lista [i,j], a lista [j,i] também está presente na lista. Caso não esteja, adiciona um ao contador
for a in relation_list:
  if [a[1],a[0]] not in relation_list:
    cont += 1
print('Total de arestas não recíprocas:')
print(cont)
```
