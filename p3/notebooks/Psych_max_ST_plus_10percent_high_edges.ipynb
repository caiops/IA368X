# Obtendo a rede de condições psiquiátricas para melhor visualização

Essa rede foi obtida através da árvore geradora máxima, adicionando a ela os 10% das arestas com os pesos mais altos.

```python
import pandas as pd
import numpy as np

#Leitura dos dados
psych = pd.read_csv('https://github.com/caiops/IA368X/raw/main/p3/data/processed/network_psych.csv')
psych_max_st = pd.read_csv('https://github.com/caiops/IA368X/raw/main/p3/data/interim/network_psych_Kruskal_Max_Spanning_Tree.csv')
```

```python
#Visualizando a tabela de arestas da árvore geradora máxima
psych_max_st
```

```python
#Quebrando a coluna "name" de modo a separar os nós
new = psych_max_st["name"].str.split(expand = True)
```

```python
#Visualizando o dataframe gerado
new
```

```python
#Criando um novo dataframe contendo os nós e os valores correspondentes de log odds para a árvore geradora máxima
psych_reduced = pd.DataFrame()
psych_reduced['i'] = new[3]
psych_reduced['j'] = new[0]
psych_reduced['log_odds'] = psych_max_st['log_odds']
```

```python
#Visualizando o dataframe gerado
psych_reduced
```

```python
#Ordenando a rede de condições psiquiátricas em ordem decrescente com base nos valores de log odds
psych_sort = psych.sort_values(by=['log_odds'], ascending=False, ignore_index=True)
#Selecionando apenas as 10% primeiras linhas (correspondentes às 10% arestas com maior log odds)
psych_sort_reduced = psych_sort.loc[0:(int(len(psych_sort)*0.1)-1)]
```

```python
#Visualizando a tabela obtida
psych_sort_reduced
```

```python
#Concatenando as tabelas da árvore geradora máxima e das 10% arestas com maior log odds
psych_vis = pd.concat([psych_reduced, psych_sort_reduced], ignore_index=True)
psych_vis
```

```python
#Removendo possíveis colunas duplicadas
psych_vis.drop_duplicates(inplace=True, ignore_index=True)
```

```python
#Visualizando a rede de condições psiquiátricas reduzida, gerada para facilitar a visualização
psych_vis
```

O dataframe gerado foi então salvo como um arquivo .csv.

```python
psych_vis.to_csv('network_psych_vis.csv', index = False)
```
